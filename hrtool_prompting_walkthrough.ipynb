## 4. Running the Streamlit Application

After building the app structure with AI assistance, the final step is to run the Streamlit application and ensure all dependencies are properly installed.

### Installing Dependencies

Before running the app, ensure all required packages are installed. The project includes a `requirements.txt` file that lists all necessary dependencies:
```
pip install -r requirements.txt
```
### Running the Streamlit App

Streamlit applications cannot be run like regular Python scripts. Instead, use the `streamlit run` command:
```
streamlit run app.py
```
This command will:
1. Start a local Streamlit server
2. Automatically open your default web browser to `http://localhost:8501`
3. Display the interactive HR Tool dashboard

### Important Notes

- Do not run the app with `python app.py` - this will generate errors about missing ScriptRunContext
- The app will automatically reload when you save changes to the code files
- To stop the server, press `Ctrl+C` in the terminal
- If port 8501 is already in use, Streamlit will automatically select the next available port

### Troubleshooting

If you encounter issues:
1. Data not loading: ensure the path in `utils.py` correctly points to your CSV file
2. Import errors: verify all packages in `requirements.txt` are installed
